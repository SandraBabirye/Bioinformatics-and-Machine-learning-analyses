```python
# Define the list of drug resistance genes
efflux_pumps = ['Rv1258c','Rv0194','emrB', 'Rv2025c',
                'Rv3041c', 'mkl', 'Rv1463', 'Rv2326c', 'Rv3197'
                'Rv1747','Rv0986', 'Rv0987','Rv2687c', 'Rv1456c', 'Rv2687c'
                'drrB','dppD','dppA','Rv1273c'
                'subI','cysW'
]
# Assuming 'cds_data' is your DataFrame

# Subset the DataFrame based on the known list of genes
subset_data = low[low['Gene_id'].isin(efflux_pumps)].copy()

# Add a new column 'Category' indicating 'Drug Resistance'
subset_data['Category'] = 'Efflux_pump'

# Reset index if needed
subset_data.reset_index(drop=True, inplace=True)

# Print the subsetted DataFrame
subset_data.tail()
subset_data.to_csv("Low_CD4_results/Efflux_pumps.genes.csv")
```

```python
# Define the list of drug resistance genes
Virulence_genes = ['Rv1794','ecc', 'esp', 'esx','fbp', 'hbhA','hspX','icl',
             'ideR'	,'irtA','irtB', 'lipF','mbtA', 'btJ','mbt'	
             'myc','pho', 'relA','erp','mce'
]

#Filter genes that contain any of the names in Virulence_genes
subset_df = low[low['Gene_id'].str.contains('|'.join(Virulence_genes))]

subset_df['Gene_id'].value_counts().to_csv("Low_CD4_results/virulence_gene_counts_low.csv")
subset_df.to_csv("Low_CD4_results/virulence_low_genes.csv")
```
